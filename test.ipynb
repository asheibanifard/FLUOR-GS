```python
# ── Load the original volume ──────────────────────────────────────
vol_path = "dataset/10-2900-control-cell-05_cropped_corrected.tif"
vol_np   = load_volume(vol_path)          # numpy (Z, Y, X) in [0, 1]
print(f"Volume shape: {vol_np.shape}  dtype: {vol_np.dtype}")

# ── Top-down MIP (Z-axis projection) in grayscale ────────────────
mip_z = vol_np.max(axis=0)   # project along Z → (Y, X)

plt.figure(figsize=(8, 8))
plt.imshow(mip_z, cmap="gray")
plt.title("MIP – Z axis (top-down)")
plt.axis("off")
plt.tight_layout()
plt.show()
```
